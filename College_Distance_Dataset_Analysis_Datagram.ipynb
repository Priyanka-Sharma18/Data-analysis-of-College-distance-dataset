# Replacing Score by MEAN

```python
mean_score = df["score"].astype("float").mean(axis=0)
print(mean_score)
df["score"].replace(np.nan, mean_score , inplace=True)
```

```python
missing_data = df.isnull()
print (missing_data["score"].value_counts())
```

# Replacing Income  And Urban by most frequent data (Mode)

```python
df["income"].replace(np.nan, "low", inplace=True)
missing_data = df.isnull()
print (missing_data["income"].value_counts())
```

```python
df["urban"].replace(np.nan, "no", inplace=True)
missing_data = df.isnull()
print (missing_data["urban"].value_counts())
```

```python
df['income'].value_counts()
```

```python
df['urban'].value_counts()
```

# For wage - Replacing the values with the mean when grouped by tuition 
# Analysed as below 

```python
df[['tuition','wage']].head(5)
```

```python
df[['tuition','wage']][30:34]
```

```python
df[['tuition','wage']][138:144]
```

```python
# Replaced in filled_wage
df['filled_wage'] = df.groupby('tuition')['wage'].transform(
    lambda grp: grp.fillna(np.mean(grp)))
```

```python
#Drop wage
df.drop("wage",axis=1,inplace=True)
df.head(3)
```

## <B> Education vs Score</B>

```python
pearson_coef, p_value = stats.pearsonr(df['score'], df['education'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```

## Conclusion:

Since the p-value is < 0.001, the correlation between score and education is statistically significant, and the linear relationship is also defined (~0.46420).

## Analysing Some Combined Values 
#### By binning

```python
bins = np.linspace(min(df["score"]), max(df["score"]), 4)
bins
group_names = ['Low', 'Medium', 'High']
df['score-binned'] = pd.cut(df['score'], bins, labels=group_names, include_lowest=True )
#df[['score','score-binned']].head(20)
df["score-binned"].value_counts()
```

```python
bins = np.linspace(min(df["tuition"]), max(df["tuition"]), 4)
bins
group_names = ['Low', 'Medium', 'High']
df['tuition-binned'] = pd.cut(df['tuition'], bins, labels=group_names, include_lowest=True )
df["tuition-binned"].value_counts()
```

```python
bins = np.linspace(min(df["filled_wage"]), max(df["filled_wage"]), 4)
bins
group_names = ['Low', 'Medium', 'High']
df['wage-binned'] = pd.cut(df['filled_wage'], bins, labels=group_names, include_lowest=True )
df["wage-binned"].value_counts()
```
